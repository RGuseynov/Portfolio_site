## Traitement des données

***Import des modules nécessaires.***

```python
import pandas as pd
import numpy as np
import json
```

***Chargement d'une année de données à partir des CSV compressés. Chaque station a son propre CSV. Ensemble ces CSV sont compressés dans une même archive pour chaque année. Je prends l'archive 2000.tar.gz ici. <br>
Dans ce notebook, je vais montrer comment traiter une année, je mettrais à la fin la fonction pour traiter et réunir l'ensemble des années.***

```python
folderPath = "../../../data/climat/daily_raw"
year = 2000

df = pd.read_csv(f"{folderPath}/{year}.tar.gz", compression='gzip')
df
```

***Lors du chargement des CSV, le nom des CSV et chacun de leur headers ont étaient chargés commes des entrés(lignes).<br>
-Je vais remédier à ça en spécifiant bien que la première colonne correspond à l'id des stations. <br>
-Je garde que les entrées qui ne sont pas des headers en utilisant la colonne DATE et en gardant que les lignes qui contiennent une date et non "DATE" littéralement. <br>
-Finalement, j'efface les entrées qui ont des valeurs manquantes car d'après les spécifications des données source, les valeurs manquantes sont censées être taguées. (On verra un peu plus loin pour ces tags). <br>
Ceci fait, on passe de 2555253 entrées à 2539330 entrées.***

```python
df = df.rename({df.columns[0]: 'STATION'}, axis=1)
df = df[df["DATE"] != "DATE"]
df = df.dropna()
df
```

```python
df.info()
```

***Pour alléger un peu notre dataframe et pouvoir travailler avec les valeurs numériques et les dates, on va procéder à quelques conversions de types***

```python
float_columns = ["LATITUDE", "LONGITUDE", "ELEVATION", "TEMP", "MAX", "MIN", 
                "DEWP", "VISIB", "WDSP", "MXSPD", "PRCP", "SNDP", "STP", "SLP", "GUST", "VISIB"]
df[float_columns] = df[float_columns].astype(np.float32)
df["STATION"] = df["STATION"].astype(np.int64)
df["DATE"] = pd.to_datetime(df["DATE"])
df.info()
```

***Pour séléctionner les pays d'Europe, je charge un json qui contient les corréspondances entre Pays et l'ancien code FIPS.***

```python
country_listPath = "../../../data/climat/country_list.json"
with open(country_listPath, 'rb') as file:
    country_dict = json.load(file)
country_dict
```

***Ce code FIPS est contenu en fin de nom des stations dans la colonne NAME. <br>
-Split de la colonne NAME en 2, la 1er partie contient le nom de la station qui est reaffecté à NAME, la 2e partie contient le code pays qui est affecté à une nouvelle colonne COUNTRY. <br>
-Je garde que les pays qui sont dans le dictionnaire. <br>
-Remplacement des code pays par le nom entier des pays. <br>
-Je garde que les stations dans les latitudes de l'Europe (des pays peuvent avoir des stations en dehors du contient européen) <br>
-Je ne garde pas les îles de l'Atlantique du Portugal qui ne font pas vraiment partis du climat européen.***

```python
df[["NAME", "COUNTRY"]] = df["NAME"].str.split(", ", n=2, expand=True)
df = df[df["COUNTRY"].isin([key for key in country_dict.keys()])]
df.loc[:, "COUNTRY"] = df.loc[:, "COUNTRY"].map(country_dict)
df = df[(df["LATITUDE"] > 35) & (df["LATITUDE"] < 72)]
df = df[~((df["COUNTRY"] == "Portugal") & (df["LONGITUDE"] < -15))]
```

```python
df["COUNTRY"].value_counts()
```

***D'après les spécifications des données sources, 9999.9 et 999.9 sont attribués à des valeurs manquantes (si on les laisse tel quel, elles vont rendre obsolète les mesures statistiques). Je les remplace donc par des valeurs manquantes. <br>
Pour les colonnes SNDP et PRCP qui correspondent à la quantité de neige et de pluie tombée, les valeurs 999.9 et 99.9 peuvent aussi signifier une abscence de neige ou de pluie. Je décide de les remplacer par 0 plutôt que par NAN.***

```python
    df[["TEMP", "DEWP", "MAX", "MIN", "SLP"]] = df[["TEMP", "DEWP", "MAX", "MIN", "SLP"]].replace(9999.9, np.nan)
    df[["WDSP", "MXSPD", "STP", "GUST", "VISIB"]] = df[["WDSP", "MXSPD", "STP", "GUST", "VISIB"]].replace(999.9, np.nan)
    df["SNDP"] = df["SNDP"].replace(999.9, 0)
    df["PRCP"] = df["PRCP"].replace(99.9, 0)
    df
```

***Après ce traitement, je souhaites savoir combien de valeurs sont manquantes dans les données (NAN).***

```python
df.isna().sum()
```

***Avec autant de valeurs manquantes pour les colonnes SLP, STP, VISIB et GUST. Plutôt que de faire des assomptions, je décide des les supprimer.
Note: il existe diverse techniques statistiques pour "boucher" ces valeurs manquantes pour essayer d'utiliser au maximum les jeux de données disponible. Elles ont leur avantages et inconvénients, je n'en ferais pas usage durant ce projet.***

```python
df = df.drop(["SLP", "STP", "VISIB", "GUST"], axis=1)
```

***La colonne FRSHTT contient la présence(1) ou l'abscence(0) des phénomènes suivants : Fog, Rain, Snow, Hail, Thunder, Tornado.***

```python
df["FRSHTT"]
```

***Extract (ou unpack) de ces valeurs dans de nouvelles colonnes respectives.***

```python
df[["FOG", "RAIN", "SNOW", "HAIL", "THUN", "TORN"]] = df["FRSHTT"].str.extract(r"(.)(.)(.)(.)(.)(.)")
df[["FOG", "RAIN", "SNOW", "HAIL", "THUN", "TORN"]] = df[["FOG", "RAIN", "SNOW", "HAIL", "THUN", "TORN"]].astype(np.int8)
df[["FOG", "RAIN", "SNOW", "HAIL", "THUN", "TORN"]]
```

***Je regarde la moyenne de ces colonnes pour avoir la fréquence à laquel ces événements se produisent/sont enregistrés en Europe.
Je constate l'extrême rareté des tornades, je décide de supprimer cette colonne du fait de cette marginalité et par le fait qu'elle risque d'être source de désiquilibre et d'outlier durant le machine learning***

```python
df[["FOG", "RAIN", "SNOW", "HAIL", "THUN", "TORN"]].mean()
```

```python
df = df.drop("TORN", axis=1)
```

***Conversion des mesures dans le format international***

```python
# conversion Fahrenheit en Celsius, DWEP = point de rose
df[["TEMP", "MAX", "MIN", "DEWP"]] = (df[["TEMP", "MAX", "MIN", "DEWP"]] -32) * 5/9
# knots en kilometre par heure
df[["WDSP", "MXSPD"]] = df[["WDSP", "MXSPD"]] * 1.852
# inche en millimetre
df["SNDP"] = df["SNDP"] * 2.54
# inche and hundredths en millimetre
df["PRCP"] = df["PRCP"] * 0.254
```

***Sélection des colonnes que je vais sauvegarder pour le reporting et le machine learning***

```python
df = df[["DATE", "STATION", "COUNTRY", "LATITUDE", "LONGITUDE", "ELEVATION", "NAME", "TEMP", "MAX", "MIN",
        "DEWP", "WDSP", "MXSPD", "PRCP", "SNDP", "FOG", "RAIN", "SNOW", "HAIL", "THUN"]]
df
```

***La prochaine étape est d'aggréger les données par mois et par station, de manière à avoir les données mensualisées. Celles-ci seront utilisés comme features pour le machine learning, et pour le reporting je suis limité à 1 GO de données pour la version gratuite de power BI.
<br>Lors d'un traitement précédent, j'ai constaté que STATION ne fesait pas un bon id, il changeait parfois dans le temps pour une même station, de même que NAME et parfois même la LATITUDE et la LONGITUDE(qui peut correspondre à un déplacement des capteurs de plus d'un kilomètre). 
<br>Je décide de chainer 3 groupement, la première par l'id STATION.<br> La 2e par NAME et COUNTRY(COUNTRY en plus pour éviter de fusionner 2 stations qui auraient le même noms dans 2 pays différents même si c'est peu probable).<br> La 3e par LATITUDE et LONGITUDE. <br>Ceci va me permettre d'unifier les stations qui ont changés de id STATIONS et/ou NAME et/ou LATITUDE LONGITUDE au cours d'un mois ou des anomalies mais qui correspondent à la même entité géographique. <br>Plus tard, en bas de ce notebook, je reactualiserais les id STATIONS, NAME, LATITUDE et LONGITUDE pour l'ensemble des mois et des années pour correspondre au plus récents.***

***Aggregation par mois et par id STATION: <br>
-NAME, COUNTRY, LATITUDE, LONGITUDE, ELEVATION: je garder la dernière valeur du mois (la plus à jour donc si changemet au cours du mois) <br>
-DAYS_WITH_MEASURES: nouvelle colonne que je créer pour savoir combien de jours ont étés aggreger pour chaque groupe <br>
-TEMP, DEWP, WDSP: je veut la moyenne mensuel de ces valeurs <br>
-MAX, MIN, MXSPD: je veux les valeurs extême du mois pour ces valeurs <br>
-SNDP, PRCP: je veux la quantité total neige et de pluie par mois <br>
FOG, RAIN, SNOW, HAIL, THUN: je veux le nombe de jours total dans le mois où ces phénomènes ont été observés***

```python
df = df.groupby([df["DATE"].dt.to_period('m'), df["STATION"]]).agg(        
    NAME=("NAME","last"), COUNTRY=("COUNTRY","last"), 
    LATITUDE=("LATITUDE","last"), LONGITUDE=("LONGITUDE","last"), 
    ELEVATION=("ELEVATION","last"), DAYS_WITH_MEASURES=('TEMP','count'), TEMP=('TEMP','mean'), 
    MAX=("MAX",'max'), MIN=("MIN",'min'), DEWP=("DEWP",'mean'), WDSP=("WDSP",'mean'), 
    MXSPD=("MXSPD",'max'), SNDP=("SNDP",'sum'), PRCP=("PRCP",'sum'), FOG=("FOG",'sum'), 
    RAIN=("RAIN",'sum'), SNOW=("SNOW",'sum'), HAIL=("HAIL",'sum'), THUN=("THUN",'sum'))
df
```

```python
df = df.reset_index()
df
```

***2e groupement: par NAME et COUNTRY***  

```python
df = df.groupby([df["DATE"], df["NAME"], df["COUNTRY"]]).agg(        
    STATION=("STATION","last"), LATITUDE=("LATITUDE","last"), LONGITUDE=("LONGITUDE","last"), 
    DAYS_WITH_MEASURES=("DAYS_WITH_MEASURES","sum"), ELEVATION=("ELEVATION","last"), TEMP=('TEMP','mean'), 
    MAX=("MAX",'max'), MIN=("MIN",'min'), DEWP=("DEWP",'mean'), WDSP=("WDSP",'mean'), 
    MXSPD=("MXSPD",'max'), SNDP=("SNDP",'mean'), PRCP=("PRCP",'mean'), FOG=("FOG",'mean'), 
    RAIN=("RAIN",'mean'), SNOW=("SNOW",'mean'), HAIL=("HAIL",'mean'), THUN=("THUN",'mean'))
df
```

***3e groupement: par LATITUDE ET LONGITUDE***

```python
df = df.reset_index()
df = df.groupby([df["DATE"], df["LATITUDE"], df["LONGITUDE"]]).agg(        
    STATION=("STATION","last"), NAME=("NAME","last"), COUNTRY=("COUNTRY","last"), 
    DAYS_WITH_MEASURES=("DAYS_WITH_MEASURES","sum"), ELEVATION=("ELEVATION","last"), TEMP=('TEMP','mean'), 
    MAX=("MAX",'max'), MIN=("MIN",'min'), DEWP=("DEWP",'mean'), WDSP=("WDSP",'mean'), 
    MXSPD=("MXSPD",'max'), SNDP=("SNDP",'mean'), PRCP=("PRCP",'mean'), FOG=("FOG",'mean'), 
    RAIN=("RAIN",'mean'), SNOW=("SNOW",'mean'), HAIL=("HAIL",'mean'), THUN=("THUN",'mean'))
df
```

***Pour avoir des données mensuels le plus fiable possible, je garde que les groupes (mois/station) qui ont  au moins 25 jours de mesures dans le mois, je supprime ensuite cette colonne.***

```python
df = df[df["DAYS_WITH_MEASURES"] >= 25]
df = df.drop("DAYS_WITH_MEASURES", axis=1)
```

***La fonction final de process d'une année:***

```python
def process_a_year(folderPath:str, year:int) -> pd.DataFrame:
    df = pd.read_csv(f"{folderPath}/{year}.tar.gz", compression='gzip')
    df = df.rename({df.columns[0]: 'STATION'}, axis=1)
    df = df[df["DATE"] != "DATE"]
    df = df.dropna()
    
    # types conversion
    float_columns = ["LATITUDE", "LONGITUDE", "ELEVATION", "TEMP", "MAX", "MIN", 
                    "DEWP", "VISIB", "WDSP", "MXSPD", "PRCP", "SNDP"]
    df[float_columns] = df[float_columns].astype(np.float32)
    df["STATION"] = df["STATION"].astype(np.int64)
    df["DATE"] = pd.to_datetime(df["DATE"])

    country_listPath = "data/climat/country_list.json"
    with open(country_listPath, 'rb') as file:
        country_dict = json.load(file)
    df[["NAME", "COUNTRY"]] = df["NAME"].str.split(", ", n=2, expand=True)
    df = df[df["COUNTRY"].isin([key for key in country_dict.keys()])]
    df.loc[:, "COUNTRY"] = df["COUNTRY"].map(country_dict)
    df = df[(df["LATITUDE"] > 35) & (df["LATITUDE"] < 72)]
    df = df[~((df["COUNTRY"] == "Portugal") & (df["LONGITUDE"] < -15))]

    df[["TEMP", "DEWP", "MAX", "MIN", "SLP"]] = df[["TEMP", "DEWP", "MAX", "MIN", "SLP"]].replace(9999.9, np.nan)
    df[["WDSP", "MXSPD", "STP", "GUST", "VISIB"]] = df[["WDSP", "MXSPD", "STP", "GUST", "VISIB"]].replace(999.9, np.nan)
    df["SNDP"] = df["SNDP"].replace(999.9, 0)
    df["PRCP"] = df["PRCP"].replace(99.9, 0)  
    df[["FOG", "RAIN", "SNOW", "HAIL", "THUN", "TORN"]] = df["FRSHTT"].str.extract(r"(.)(.)(.)(.)(.)(.)")
    df[["FOG", "RAIN", "SNOW", "HAIL", "THUN", "TORN"]] = df[["FOG", "RAIN", "SNOW", "HAIL", "THUN", "TORN"]].astype(np.int8)
    
    # sélection des colonnes
    df = df[["DATE", "STATION", "COUNTRY", "LATITUDE", "LONGITUDE", "ELEVATION", "NAME", "TEMP", "MAX", "MIN",
            "DEWP", "WDSP", "MXSPD", "PRCP", "SNDP", "FOG", "RAIN", "SNOW", "HAIL", "THUN"]]

    # conversion Fahrenheit en Celsius, DWEP = point de rose
    df[["TEMP", "MAX", "MIN", "DEWP"]] = (df[["TEMP", "MAX", "MIN", "DEWP"]] -32) * 5/9
    # knots en kilometre par heure
    df[["WDSP", "MXSPD"]] = df[["WDSP", "MXSPD"]] * 1.852
    # inche en millimetre
    df["SNDP"] = df["SNDP"] * 2.54
    # inche and hundredths en millimetre
    df["PRCP"] = df["PRCP"] * 0.254

    # aggregation
    df = df.groupby([df["DATE"].dt.to_period('m'), df["STATION"]]).agg(        
        NAME=("NAME","last"), COUNTRY=("COUNTRY","last"), 
        LATITUDE=("LATITUDE","last"), LONGITUDE=("LONGITUDE","last"), 
        ELEVATION=("ELEVATION","last"), DAYS_WITH_MEASURES=('TEMP','count'), TEMP=('TEMP','mean'), 
        MAX=("MAX",'max'), MIN=("MIN",'min'), DEWP=("DEWP",'mean'), WDSP=("WDSP",'mean'), 
        MXSPD=("MXSPD",'max'), SNDP=("SNDP",'sum'), PRCP=("PRCP",'sum'), FOG=("FOG",'sum'), 
        RAIN=("RAIN",'sum'), SNOW=("SNOW",'sum'), HAIL=("HAIL",'sum'), THUN=("THUN",'sum'))
    df = df.reset_index()
    
    df = df.groupby([df["DATE"], df["NAME"], df["COUNTRY"]]).agg(        
        STATION=("STATION","last"), LATITUDE=("LATITUDE","last"), LONGITUDE=("LONGITUDE","last"), 
        DAYS_WITH_MEASURES=("DAYS_WITH_MEASURES","sum"), ELEVATION=("ELEVATION","last"), TEMP=('TEMP','mean'), 
        MAX=("MAX",'max'), MIN=("MIN",'min'), DEWP=("DEWP",'mean'), WDSP=("WDSP",'mean'), 
        MXSPD=("MXSPD",'max'), SNDP=("SNDP",'mean'), PRCP=("PRCP",'mean'), FOG=("FOG",'mean'), 
        RAIN=("RAIN",'mean'), SNOW=("SNOW",'mean'), HAIL=("HAIL",'mean'), THUN=("THUN",'mean'))
    df = df.reset_index()
    
    df = df.groupby([df["DATE"], df["LATITUDE"], df["LONGITUDE"]]).agg(        
        STATION=("STATION","last"), NAME=("NAME","last"), COUNTRY=("COUNTRY","last"), 
        DAYS_WITH_MEASURES=("DAYS_WITH_MEASURES","sum"), ELEVATION=("ELEVATION","last"), TEMP=('TEMP','mean'), 
        MAX=("MAX",'max'), MIN=("MIN",'min'), DEWP=("DEWP",'mean'), WDSP=("WDSP",'mean'), 
        MXSPD=("MXSPD",'max'), SNDP=("SNDP",'mean'), PRCP=("PRCP",'mean'), FOG=("FOG",'mean'), 
        RAIN=("RAIN",'mean'), SNOW=("SNOW",'mean'), HAIL=("HAIL",'mean'), THUN=("THUN",'mean'))
    df = df.reset_index()

    df = df[df["DAYS_WITH_MEASURES"] >= 25]
    df = df.drop("DAYS_WITH_MEASURES", axis=1)
   
    return df
```

***La fonction final de process de plusieurs années:***

```python
def process_years(folderPath:str, begin_year:int, end_year:int, destinationPath:str):
    df_list = []
    # appel de la fonction process_a_year pour chaque année et concaténation des résultats dans un unique dataframe
    for year in range(begin_year, end_year + 1):
        df = process_a_year(folderPath, year)
        df_list.append(df)
    dfs = pd.concat(df_list, ignore_index=True)
   
    # on maj les valeurs d'identification sur l'ensemble des annees pour correspondre au données les plus récentes
    dfs[["NAME", "COUNTRY", "LATITUDE", "LONGITUDE", "ELEVATION"]] = dfs[["NAME", "COUNTRY", "STATION", 
        "LATITUDE", "LONGITUDE", "ELEVATION"]].groupby(["STATION"]).transform('last')   
    # sert dans le cas où l'id STATION a changés dans le temps
    dfs[["STATION", "LATITUDE", "LONGITUDE", "ELEVATION"]] = dfs[["NAME", "COUNTRY", "STATION", 
        "LATITUDE", "LONGITUDE", "ELEVATION"]].groupby(["NAME", "COUNTRY"]).transform('last')
    # sert dans le cas ou une station(même emplacement en LAT et LONG) a changé d'id STATION et de NAME
    dfs[["NAME", "COUNTRY", "STATION", "ELEVATION"]] = dfs[["NAME", "COUNTRY", "STATION", 
        "LATITUDE", "LONGITUDE", "ELEVATION"]].groupby(["LATITUDE", "LONGITUDE"]).transform('last')

    # nombre de mois avec des mesures par station
    months_with_measures = dict(dfs["STATION"].value_counts())
    dfs["MONTHS_WITH_MEASURES"] = dfs["STATION"].map(months_with_measures)
    # on drop les stations qui ont moins de 6 mois de mesures pour plus de fiabilité dans les données
    dfs_drop = dfs[dfs["MONTHS_WITH_MEASURES"] < 6]
    dfs = dfs[dfs["MONTHS_WITH_MEASURES"] >= 6]

    # on créer un dataframe qui contient que les informations géogarphique d'un côté et 
    # un dataframe qui contientl'ensemble des mesures de l'autre
    dfs_geo_dim = dfs[["STATION", "NAME", "COUNTRY", "LATITUDE", "LONGITUDE", "ELEVATION"]].drop_duplicates("STATION")
    dfs_fact = dfs.drop(["NAME", "COUNTRY", "LATITUDE", "LONGITUDE", "ELEVATION",
        "MONTHS_WITH_MEASURES"], axis=1)
    
    # on sauvegarde les données dans 2 CSV 
    dfs_fact.to_csv(f"{destinationPath}/ClimatFACT.csv", index=False)
    dfs_geo_dim.to_csv(f"{destinationPath}/StationDIM.csv", index=False)
```
