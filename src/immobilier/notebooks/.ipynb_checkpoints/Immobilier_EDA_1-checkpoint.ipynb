***Import du module gzip pour l'ouverture des fichiers compressés et de la librairie pandas. Changement d'une option d'affichage pour éviter d'avoir les nombre sous format scientifique dans les pandas dataframe.***

```python
import gzip
import pandas as pd

pd.set_option('display.float_format', lambda x: '%.5f' % x)
```

```python
folder_path = "../../../data/immobilier/transactions_raw"
with gzip.open(f'{folder_path}/full{2019}.csv.gz', 'rb') as f:
            df = pd.read_csv(f)
```

***Utilisation de la librairie pandas_profiling pour une analyse automatique sur le jeu de données.
Sélection que de la corrélation de Pearson pour la partie correlation afain de diminuer le temps d'analyse.***

```python
from pandas_profiling import ProfileReport
profile = ProfileReport(df, title="DVF 2019 raport initial", html={"inline":False}, 
                        correlations={
                             "pearson": {"calculate": True},
                             "spearman": {"calculate": False},
                             "kendall": {"calculate": False},
                             "phi_k": {"calculate": False},
                             "cramers": {"calculate": False},
                         })
profile.to_file("../../../local_files/DVF_2019_raport_initial.html")
```
